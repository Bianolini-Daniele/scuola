```python
#es 1

def valori(lista):
  n_mag = lista[1]
  n_min = lista[1]
  somma = 0
  media = 0
  for i in range(len(lista)):
    if lista[i] > n_mag:
      n_mag = lista[i]
    if lista[i] < n_min:
      n_min = lista[i]
    somma += lista[i]
  media = somma/len(lista)
  return n_mag, n_min, somma, media


lista = [1,2,3,4,5,6,7,8]
n,p,s,m = valori(lista)
print(n,",",p,",",s,",",m)

```

```python
#es 3

def primi(lista):
    def is_primo(n):
        if n < 2:
            return False
        if n == 2:
            return True
        if n % 2 == 0:
            return False
        for i in range(3, int(n**0.5) + 1, 2):
            if n % i == 0:
                return False
        return True
    return [n for n in lista if is_primo(n)]

numeri = [1, 2, 3, 4, 5, 6, 7, 15, 17, 20]
print(primi(numeri))
```

```python
#es 4
def quadrati(N):
    if N < 1:
        return 0

    somma = 0
    for i in range(1, N + 1):
        quadrato = i ** 2
        somma += quadrato

    return somma

N_valore = 5
risultato_quadrati = quadrati(N_valore)

print(f"Calcolo la somma dei quadrati dei primi {N_valore} numeri (1^2 + 2^2 + 3^2 + 4^2 + 5^2)")
print(f"Risultato: **{risultato_quadrati}**")

```
